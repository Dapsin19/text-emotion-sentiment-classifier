```python
from transformers import pipeline

# Load pretrained sentiment classifier fine-tuned on SST-2
classifier = pipeline("text-classification", model="distilbert-base-uncased-finetuned-sst-2-english")

# Example texts
texts = [
    "I love this product!",
    "This is terrible and disappointing.",
    "It was okay, not great but not bad."
]

# Run predictions
for text in texts:
    result = classifier(text)[0]
    print(f"Text: {text}")
    print(f"Label: {result['label']} | Score: {result['score']:.4f}")
    print("-" * 40)
```

```python
from transformers import pipeline

# Load the emotion classification model
emotion_classifier = pipeline("text-classification",
                              model="j-hartmann/emotion-english-distilroberta-base",
                              return_all_scores=True)

# Sample texts
texts = [
    "I’m so happy with the results!",
    "This is making me really angry.",
    "I feel empty and lost.",
    "Wow, I wasn’t expecting that!",
    "I love the way this turned out."
]

# Get predictions
for text in texts:
    result = emotion_classifier(text)[0]
    # Sort by score descending
    sorted_result = sorted(result, key=lambda x: x['score'], reverse=True)
    top_emotion = sorted_result[0]

    print(f"Text: {text}")
    print(f"Emotion: {top_emotion['label']} | Confidence: {top_emotion['score']:.4f}")
    print("-" * 40)
```

```python
import re

def clean_text(text):
    text = text.lower()                                # Lowercase
    text = re.sub(r"http\S+|www.\S+", "", text)         # Remove URLs
    text = re.sub(r"[^a-zA-Z\s]", "", text)             # Remove special chars/numbers
    text = re.sub(r"\s+", " ", text).strip()            # Remove extra spaces
    return text
```

```python
from transformers import pipeline
import re

# Emotion classifier
emotion_classifier = pipeline("text-classification",
                              model="j-hartmann/emotion-english-distilroberta-base",
                              return_all_scores=True)

# Text cleaning function
def clean_text(text):
    text = text.lower()
    text = re.sub(r"http\S+|www.\S+", "", text)
    text = re.sub(r"[^a-zA-Z\s]", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

# Input texts
texts = [
    "I’m so happy with the results! 🎉",
    "This is making me really angry 😡!!",
    "I feel empty and lost... http://sad.com",
    "Wow, I wasn’t expecting that 😲",
    "I love the way this turned out 💖"
]

# Loop with cleaning + prediction
for text in texts:
    cleaned = clean_text(text)
    result = emotion_classifier(cleaned)[0]
    top_emotion = sorted(result, key=lambda x: x['score'], reverse=True)[0]

    print(f"Original: {text}")
    print(f"Cleaned: {cleaned}")
    print(f"Predicted Emotion: {top_emotion['label']} | Score: {top_emotion['score']:.4f}")
    print("-" * 50)
```
